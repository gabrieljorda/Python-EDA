### Teste do Qhi-Quadrado

- H0 (Hipótese Nula): As duas variáveis são independentes.
- H1 (Hipótese Complementar): As duas veriáveis não são independentes.
- No teste, o objetibo é confirmar ou recusar a hipótese nula
- Quando o valor de probabilidade H0 for menor que 0.05 (p-value) recusamos a hipótese nula e aceitamos a complementar.



```python
df_tabcont = pd.crosstab(df_churn["Churn"],df_churn["Contract"] )
```

```python
from scipy.stats import chi2_contingency
```

```python
chi_score =  chi2_contingency(df_tabcont)
```

```python
chi_score
```

```python
# Número decimal de uma notação cientifica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
score_churn = pd.Series(chi_score[0])
pcalues_churn = pd.Series(chi_score[1])
```

```python
df_chi_churn = pd.DataFrame({
    "qui2": score_churn,
    "pvalue": pcalues_churn
})
```

```python
df_chi_churn
```

As variáveis não são independentes. E pelo chi² alta correlação entre as variaveis 

### Correlação entre variável qualitativa e quantitativa 

Hipótese: cliente com menos de 6 meses de contrato é mais propenso ao churn

```python
import numpy as np
```

```python
df_churn["TempoMenorq6"] = np.where(df_churn["tenure"] < 6, "Sim", "Não")
```

```python
df_churn.head()
```

```python
df_costrab_tempo  = pd.crosstab(df_churn["Churn"], df_churn["TempoMenorq6"], margins=True, margins_name="Total")
df_costrab_tempo
```

```python
chi_scores = chi2_contingency(df_costrab_tempo)
```

```python
chi_score
```

```python
scores_churn = pd.Series(chi_scores[0])
pvalues_churn = pd.Series(chi_scores[1])
```

```python
df_chi_churn = pd.DataFrame({
    "chi2": scores_churn,
    "pvalue": pvalues_churn
})
```

```python
df_chi_churn
```

As duas variáveis não são independentes. E há uma correlação entre as variáveis, mas não é tão forte quanto a variável tipo de contrato.

### Técnica IQR
- 1° quartil e 3° quartil | 3° quartil - 1° quartil

```python
q1_total = df_churn_month["TotalCharges"].quantile(0.25)
```

```python
q3_total = df_churn_month["TotalCharges"].quantile(0.75)
```

```python
iqr = q3_total - q1_total
iqr
```

```python
lim_inf = q1_total - (1.5 * iqr)
lim_sup = q3_total + (1.5 * iqr)
```

```python
lim_inf , lim_sup
```

```python
x1 = len(df_churn_month[(df_churn_month["TotalCharges"] < lim_inf) | (df_churn_month["TotalCharges"] > lim_sup)])
x1
```
